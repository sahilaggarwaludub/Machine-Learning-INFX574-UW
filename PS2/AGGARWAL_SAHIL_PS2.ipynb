```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return -(x*x)

def gradient(x):
    return (-2)*x
```

```python
import time
def one_d_gradient_ascent(x0,maxIter,R,epsilon=0.0001):
    i=0
    while i<maxIter:
        print ("x value:",x0)
        print ("Function value:", f(x0))
        print ("Gradient of the function", gradient(x0),'\n')
        x1 = x0 + (R*gradient(x0))
        if(abs(x1-x0) <=epsilon):
            break
        x0 = x1
        i=i+1
    print("Iterations:", i)
    print("Maxima", x1)
```

```python
one_d_gradient_ascent(4,1000,0.1)
```

##### Question 7

```python
def f(x, A):
    return (-1)*x.T @ A @ x

def gradient(x, A):
    return (-1)*(A + A.T) @ x 
```

```python
def two_d_gradient_ascent(x0, A, R, max_iterations, mat_x, e=0.00001):
    start_time = time.time()
    for i in range(max_iterations):
        x1 = x0 + (R*gradient(x0, A))
        mat_x = np.vstack((mat_x, x1.T))
        if((x1-x0).any() <=e):
            break
        x0 = x1
    print("\nConverged Iterations:", i)
    print("\nMaximum of the function:", x1)
    print("\nTime taken:", time.time() - start_time)
    print("\nAll values of x:", mat_x)
    return mat_x
```

```python
x = np.matrix("2;-3")
print(x)
```

```python
A = np.matrix("1,2; 2,8")
print(A)
```

```python
mat_x =  two_d_gradient_ascent(x, A, 0.1, 10000, mat_x=x.T)
```

```python
mat_x_2 = two_d_gradient_ascent(x, A, 0.11, 10000, mat_x=x.T)
```

For the same value of x and different values of learning rate,the number of iterations and the time taken to converge is more for smaller learning rate.

```python

def gradient_ascent_contour(mat_x, A, max_iterations,lev,p1=6,p2=6):    
    n = 30
    ex1 = np.linspace(-p1, p2, num=n)
    ex2 = np.linspace(-p1, p2, num=n)
    grid1, grid2 = np.meshgrid(ex1, ex2)
    ## fill the grid via looping.  You may prefer to use ufuncs instead.
    z = np.empty_like(grid1)
    for i in range(z.shape[0]):
        for j in range(z.shape[1]):
            x = np.array([grid1[i,j], grid2[i,j]])
            z[i,j] = f(x,A)
    plt.figure(figsize=(12,12))
    p = plt.contour(grid1, grid2, z,levels = lev,
                    colors = 'black')
    plt.clabel(p, inline=1, fontsize=10)
    mat_x = np.array(mat_x)
    for i in range(100):
        plt.scatter(mat_x[i][0],mat_x[i][1], c='red', s=50)
        if(i>0):
            plt.plot([mat_x[i][0],mat_x[i-1][0]],[mat_x[i][1],mat_x[i-1][1]], c='red')
    plt.scatter(mat_x[-1][0],mat_x[-1][1], c='blue', s=50)
    plt.show()
```

```python
def normal_distribution(R, N):
    mean_of_means = np.random.normal(size=(R,N)).mean(axis=1)
    print("Mean of means:", mean_of_means.mean())
    std = np.sqrt(mean_of_means.var())
    print("Standard deviation:",std)
    x1 = np.percentile(mean_of_means, 97.5)
    x2 = np.percentile(mean_of_means, 2.5)
    points_outside = np.logical_or(mean_of_means < x2, mean_of_means > x1)
    print("confidence interval:", x2, x1)
    print("95% Confidence Region: ", points_outside.mean())
```

```python
normal_distribution(R=1000, N=10)
```

```python
normal_distribution(R=1000,N=1000)
```

```python
normal_distribution(R=1000,N=100000)
```

```python
def pareto_distribution(R, N):
    mean_of_means = np.random.pareto(a=0.5, size=(R, N)).mean(axis=1)
    print ("mean of means: ", mean_of_means.mean())
    std = mean_of_means.std()
    print ("standard dev: ", std)
    x1 = np.percentile(mean_of_means, 97.5)
    x2 = np.percentile(mean_of_means, 2.5)

    points_outside = np.logical_or(mean_of_means < x2, mean_of_means > x1)
    print("confidence interval:", x2, x1)
    print("95% Confidence Region: ", points_outside.mean())
```

```python
pareto_distribution(1000, 1000)
```

```python
pareto_distribution(1000,100)
```

```python
pareto_distribution(1000,10)
```

Pareto distributions have higher means and standard deviations than normal distributions. For k<=1, the denominator of expected value becomes very small and expected value becomes more. The expected value for k<=1 doesn't have any significance as its value is very high and unstable unlike normal distribution as these distributions are highly skewed.
